```python
#Replacing N/A
from sklearn.impute import SimpleImputer
```

```python
missingValues = SimpleImputer(missing_values = np.nan, strategy = 'mean')
```

```python
missingValues = missingValues.fit(data[['fte', 'demp']])
```

```python
data[['fte', 'demp']] = missingValues.transform(data[['fte','demp']])
```

```python
data
```

```python
data.isnull().any()
```

```python
#All N/As are transformed. 
```

```python
y = data['fte']
```

```python
x = data.iloc[:,0:3].values
```

```python
x
```

```python
import statsmodels.api as sm
```

```python
x = sm.add_constant(x)
```

```python
model1 = sm.OLS(y, x).fit()
```

```python
model1.summary(yname = 'FTE', xname = ('Intercept', 'New Jersey', 'After April 1992', 'New Jersey and after April 1992'))
```
